****We are going to EDA on Disney+Hotstar****

In first step we want to import the libraries that we want in this EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

The requried data is Import to the data by add data .Then it is Read by using the pandas libirary using read_csv()Method

```python
df=pd.read_csv('/kaggle/input/disney-hotstar-tv-and-movie-catalog/hotstar.csv')
df
```

```python
df.columns=df.columns.astype(str).str.capitalize()
df
```

```python
df.info()
```

we are going to divided the data set using the loc()method then it is saved in different variables for easy to identify

```python
Movie=df.loc[df['Type']=='movie'].reset_index(drop=True)
Movie
```

unwanted columns in the Movies is to be remove by using the drop()method 

```python
Movie.drop(['Seasons','Episodes'],axis=1,inplace=True)
```

```python
Movie
```

```python
Tv=df.loc[df['Type']=='tv'].reset_index(drop=True)
Tv
```

unwanted columns in the Tv is to be remove by using the drop()method

```python
Tv.drop('Running_time',axis=1,inplace=True)
```

```python
Tv
```

**Q1.List the Tv series by Age Rating**

```python
Tv['Age_rating'].unique()
label=['U/A 13+', 'U/A 7+', 'U/A 16+', 'U', 'A', 'PG']
Tv1=Tv['Title'].groupby(Tv['Age_rating']).size()
plt.pie(Tv1,autopct='%.2f',labels=label)
plt.show()
```

**Q2..List the Movies by Age Rating**

```python
Movie['Title'].groupby(Movie['Age_rating']).size().plot(kind='bar')
plt.show()
```

**Q3.List the yearwise released Tv Series**

```python
plt.figure(figsize=(10,8))
Tv['Title'].groupby(Tv['Year']).size().plot(kind='bar')
```

**Q4. List the yearwise released Movies**

```python

plt.figure(figsize=(16,10))
Movie['Title'].groupby(Movie['Year']).size().plot(kind='bar')
plt.show()
```
